```python
import pandas as pd
import requests
import pandas as pd
from config import x_api_key,x_api_secret_key
url='https://api-v3.receptiviti.com/v3/api/content'
headers={'X-API-KEY': x_api_key, 'X-API-SECRET-KEY':x_api_secret_key}
pd.set_option('display.max_colwidth', 1000)
```

```python
# this takes first 9999 words from each account. this may have been the issue as the text was converted tio a list
# I converted it back and rerean it. let me know if this helps.
f = pd.read_csv("data\compiled_tweets.csv")
df = pd.DataFrame(f)
df['text'] = df['text'].str[1:]
df2=df.groupby(['name'])['text'].apply(lambda x: "{%s}" % ', '.join(x))
df2=pd.DataFrame(df2)
df2['text'] = df2['text'].apply(unicodetoascii)
df2['text'] = df2['text'].apply(cleanText)
df2['text'] = df2['text'].apply(cleanText)
df3=pd.DataFrame(df[['name','party']])
df3=df3.drop_duplicates()
df3['party'] = df3.apply(lambda row: partyb (row),axis=1)
df3=df3[df3['party']!='Other']
df2 = pd.merge(df2,df3,on='name', how='left')
df2['text'] = df2['text'].str.split().str[:9998]
####just added this line to recompile text
df2['text'] = df2['text'].str.join(' ')
df2.to_csv('individual_scores.csv')
#df2.head()
```

```python
#%%
df4 = pd.DataFrame(df2.head(5))#for testing
raw_scores = []
percentile_scores = []
category_scores = []
liwc_scores = []
df2.apply(lambda r : sendLIWC(r['text'], r['party'],r['name']), axis=1)
```

```python
#%%
df_raw = pd.DataFrame(raw_scores)
df_percent = pd.DataFrame(percentile_scores)
df_cat = pd.DataFrame(category_scores)

df_raw.to_csv('liwc_individual_raw_scores2.csv',index=False)
df_percent.to_csv('liwc_individual_percentile_scores2.csv',index=False)
df_cat.to_csv('liwc_individual_categorical_scores2.csv',index=False)
```

```python
#%%
def partyb (row):
    if row['party'] == 'house-democrats':
        return 'democrats'
    if row['party'] == 'house-republicans':
        return 'republicans'
    if row['party'] == 'SenateRepublicans':
        return 'republicans'
    if row['party'] == 'SenateDemocrats':
        return 'democrats'
    return 'Other'
```

```python
#%%
import re
def cleanText(x):
    x = x.replace('RT','')
    x = x.replace(' .','.')
    x = x.replace('. ','.')
    x = re.sub('[^a-zA-Z0-9 \n\.]', '', x)
    x = re.sub(r'http\S+', '', x)
    x = re.sub(r'\w+:\s?', '', x)
    x = re.sub(r'\\xe2\\x..\\...', '', x)
    x = re.sub(r'xe2x..x..', '', x)
    x = re.sub(r'xf0x..x..x..', '', x)
    x = re.sub(r'xf0x..x..', '', x)
    x = re.sub(r'xc2...', '', x)
    x = re.sub(r'xc3xa1', 'a', x)  
    x = re.sub(r'ixc3xb3', 'i', x)
    return x
```

```python
#%%
def unicodetoascii(text):

    TEXT = (text.
            replace('\\xe2\\x80\\x99', "'").
            replace('\\xc3\\xa9', 'e').
            replace('\\xe2\\x9a\\xbe\\xef\\xb8\\x8f', "").
            replace('\\xe2\\x9c\\x94\\xef\\xb8\\x8f', "").
            replace('\\xe2\\x80\\x8a-\\xe2\\x80\\x8a', "").
            replace('\\xe2\\xac\\x87\\xef\\xb8\\x8f', "").
            replace('\\xe2\\x80\\x90', '-').
            replace('\\xe2\\x80\\x91', '-').
            replace('\\xe2\\x80\\x92', '-').
            replace('\\xe2\\x80\\x93', '-').
            replace('\\xe2\\x80\\x94', '-').
            replace('\\xe2\\x80\\x94', '-').
            replace('\\xe2\\x80\\x98', "'").
            replace('\xe2\\x80\\x99', "'").
            replace('\\xe2\\x80\\x9b', "'").
            replace('\xe2\x80\x99', "'").
            replace('\\xe2\\x80\\x9c', '"').
            replace('\\xe2\\x80\\x9c', '"').
            replace('\\xe2\\x80\\x9d', '"').
            replace('\\xe2\\x80\\x9e', '"').
            replace('\\xe2\\x80\\x9f', '"').
            replace('\\xe2\\x80\\xa6', '...').
            replace('\\xe2\\x81\\xa6', "").
            replace('\\xe2\\x81\\xa9', "").
            replace('\\xe2\\x80\\x8b', "").
            replace('\\xe2\\x80\\xb2', "'").
            replace('\\xe2\\x80\\xb3', "'").
            replace('\\xe2\\x80\\xb4', "'").
            replace('\\xe2\\x80\\xb5', "'").
            replace('\\xe2\\x80\\xb6', "'").
            replace('\\xe2\\x80\\xb7', "'").
            replace('\\xe2\\x81\\xba', "+").
            replace('\\xe2\\x81\\xbb', "-").
            replace('\\xe2\\x81\\xbc', "=").
            replace('\\xe2\\x81\\xbd', "(").
            replace('\\xe2\\x81\\xbd', "(").
            replace('\\xf0\\x9f\\x8f\\x86', "").
            replace('\\xf0\\x9f\\x93', "").
            replace('\\xe2\\x9a\\xa1', "").
            replace('\\xe2\\x86\\x92', "").
            replace('\\xe2\\x9c\\x85', "").
            replace('\\xe2\\x96\\x88', "").
            replace('\\xe2\\x9e\\xa1', "").
            replace('\\xef\\xb8\\x8f', "").
            replace('\\xe2\\x86\\x93 ', "").
            replace('\\xe2\\xac\\x87', "").
            replace('\\xe2\\x96\\xb6', "").
            replace('\\xe2\\x99\\xa5', "").
            replace('\\xf0\\x9f\\x9b\\x91', "").
            replace('\\xf0\\x9f\\x87\\xb7', "").
            replace('\\xf0\\x9f\\x87\\xba', "").
            replace('\\xe2\\x80\\xbc', "").
            replace('\\xe2\\x9d\\xa4', "").
            replace('\\xe2\\x81\\xbe', ")")   
                 )
    return TEXT

```

```python
def sendLIWC(text, party, name):
    
    print(f'sending request for userid {name}' )

    response = requests.post(url, headers=headers,json=payload(text))
    #print(text) for testing
    liwc = {}
    liwc['name'] = name
    liwc['party'] = party
    raw_score = dict(liwc)
    raw_score.update(response.json()['receptiviti_scores']['raw_scores'])
    percentile_score = dict(liwc)
    percentile_score.update(response.json()['receptiviti_scores']['percentiles'])
    category_score = dict(liwc)
    category_score.update(response.json()['liwc_scores']['categories'])
    
    raw_scores.append(raw_score)
    percentile_scores.append(percentile_score)
    category_scores.append(category_score)
```

```python
#%%
def payload(text):
    payload = {}
    tags = []
    tags.append('string')
    payload['content_tags'] = tags
    payload['content_handle'] = 'string'
    payload['language'] = 'english'
    payload['content_source'] = 4
    payload['content_date'] = '2018-12-25T00:46:05.119779'
    payload['recipient_id'] = 'string'
    payload['language_content'] = text
    return payload
```
