```python
import requests
import pandas as pd
from config import x_api_key,x_api_secret_key
url='https://api-v3.receptiviti.com/v3/api/content'
headers={'X-API-KEY': x_api_key, 'X-API-SECRET-KEY':x_api_secret_key}
```

```python
df = pd.read_csv('data/final_tweets2.csv')
df = df.drop(columns=['Unnamed: 0'])
df.reset_index(drop=True)
df.head()
```

```python
def payload(text):
    payload = {}
    tags = []
    tags.append('string')
    payload['content_tags'] = tags
    payload['content_handle'] = 'string'
    payload['language'] = 'english'
    payload['content_source'] = 4
    payload['content_date'] = '2019-01-05T00:46:05.119779'
    payload['recipient_id'] = 'string'
    payload['language_content'] = text
    return payload
```

```python
raw_scores = []
percentile_scores = []
category_scores = []
liwc_scores = []
```

```python

def sendLIWC(text, party):
    
    #print(f'sending request for userid {userid}' )

    response = requests.post(url, headers=headers,json=payload(text))
    liwc = {}
    #liwc['userid'] = userid
    liwc['party'] = party
    raw_score = dict(liwc)
    raw_score.update(response.json()['receptiviti_scores']['raw_scores'])
    percentile_score = dict(liwc)
    percentile_score.update(response.json()['receptiviti_scores']['percentiles'])
    category_score = dict(liwc)
    category_score.update(response.json()['liwc_scores']['categories'])
    
    raw_scores.append(raw_score)
    percentile_scores.append(percentile_score)
    category_scores.append(category_score)
    
    #liwc_score = dict(liwc).update(response.json()['liwc_scores']['categories'])
    
```

```python
df.apply(lambda r : sendLIWC(r['text'], r['party']), axis=1)
#test_df = df.head()
#test_df.apply(lambda r : sendLIWC(r['text'], r['party'],r['key']), axis=1)
```
